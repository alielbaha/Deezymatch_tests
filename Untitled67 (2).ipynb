```python
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer("shahrukhx01/paraphrase-mpnet-base-v2-fuzzy-matcher")
```

```python
entities = [
    "βίνσεντ",
    "zoé",
    "alain"
]
```

```python
entity_embeddings = model.encode(entities, convert_to_tensor=True)
```

```python
queries = ["vincent"]

for query in queries:
    query_embedding = model.encode(query, convert_to_tensor=True)
    similarities = util.pytorch_cos_sim(query_embedding, entity_embeddings)
    print(f"query: {query} --> sim score: {similarities}")
    best_match_idx = similarities.argmax().item()
    best_match = entities[best_match_idx]
    similarity_score = similarities[0][best_match_idx].item()

    print(f"query: {query} --> best Match: {best_match} (score: {similarity_score:.4f})")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_fscore_support, roc_auc_score, roc_curve
from sentence_transformers import SentenceTransformer, util

# Load the model
model = SentenceTransformer('shahrukhx01/paraphrase-mpnet-base-v2-fuzzy-matcher')

# Sample name pairs (true matches and false matches)
data = [
    ("Jonathan", "Johnathan", 1),
    ("Michael", "Micheal", 1),
    ("Sarah", "Sara", 1),
    ("Alex", "Alekz", 1),
    ("Robert", "Rob", 1),
    ("Sophia", "Sofia", 1),
    ("John", "Doe", 0),
    ("Alice", "Bob", 0),
    ("William", "Emma", 0),
    ("Daniel", "Olivia", 0)
]

df = pd.DataFrame(data, columns=["name1", "name2", "label"])

# Encode names using character-level tokenization
df["name1_tokenized"] = df["name1"].apply(lambda x: " ".join(x))
df["name2_tokenized"] = df["name2"].apply(lambda x: " ".join(x))

# Compute similarity scores
embeddings1 = model.encode(df["name1_tokenized"].tolist())
embeddings2 = model.encode(df["name2_tokenized"].tolist())
df["similarity_score"] = [util.cos_sim(e1, e2).item() for e1, e2 in zip(embeddings1, embeddings2)]

# Convert to NumPy arrays
labels = df["label"].values
scores = df["similarity_score"].values

# Compute Precision, Recall, F1-score
precision, recall, f1, _ = precision_recall_fscore_support(labels, scores > 0.7, average="binary")
print(f"Precision: {precision:.3f}, Recall: {recall:.3f}, F1-score: {f1:.3f}")

# Compute AUC Score
auc = roc_auc_score(labels, scores)
print(f"ROC AUC Score: {auc:.3f}")

# Plot ROC Curve
fpr, tpr, _ = roc_curve(labels, scores)
plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

# Advanced Statistical Analysis
# KS Test (Check Normality of Similarity Scores)
ks_stat, ks_p = stats.kstest(scores, 'norm')
print(f"KS Test: Statistic={ks_stat:.3f}, p-value={ks_p:.3f}")

# Mann-Whitney U Test (Difference between match & non-match scores)
match_scores = df[df["label"] == 1]["similarity_score"].values
non_match_scores = df[df["label"] == 0]["similarity_score"].values
u_stat, u_p = stats.mannwhitneyu(match_scores, non_match_scores, alternative='greater')
print(f"Mann-Whitney U Test: Statistic={u_stat:.3f}, p-value={u_p:.3f}")

# Confidence Interval for Matching Scores
conf_int = stats.norm.interval(0.95, loc=np.mean(match_scores), scale=stats.sem(match_scores))
print(f"95% Confidence Interval for Matching Scores: {conf_int}")

# Plot Similarity Score Distribution
sns.histplot(match_scores, label="Matches", color="blue", alpha=0.6, kde=True)
sns.histplot(non_match_scores, label="Non-Matches", color="red", alpha=0.6, kde=True)
plt.legend()
plt.xlabel("Similarity Score")
plt.ylabel("Frequency")
plt.title("Similarity Score Distribution")
plt.show()
```

```python
def rank_name_variations(name, variations):

    model = SentenceTransformer('shahrukhx01/paraphrase-mpnet-base-v2-fuzzy-matcher')

    name_tokenized = " ".join(name)
    variations_tokenized = [" ".join(var) for var in variations]

    name_embedding = model.encode(name_tokenized)
    variation_embeddings = model.encode(variations_tokenized)

    scores = [util.cos_sim(name_embedding, var_emb).item() for var_emb in variation_embeddings]

    ranked_variations = sorted(zip(variations, scores), key=lambda x: x[1], reverse=True)

    return ranked_variations
```

```python
name = "Jonathan"
variations = ["Johnathan", "Jon", "Johnny", "Nathan", "Jonathon"]
ranked_list = rank_name_variations(name, variations)

print('query :', name)
print('candidates:')
for var, score in ranked_list:
    print(f"{var}: {score:.4f}")
```
